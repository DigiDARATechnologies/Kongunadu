# Artificial/Deep Neural Network

```python
import numpy as np
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
dataset.head()
```

```python
# split the data into dependent and independent variable
x = dataset.iloc[:,3:-1]
y = dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
x = pd.get_dummies(x, columns=['Geography',	'Gender'])
x
```

```python
x = x.drop(['Geography_France','Gender_Female'], axis=1 )
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_stand = sc.fit_transform(x)
x_stand
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_stand, y, test_size=0.2, random_state=1)
```

# Building Deep Neural Network (MultiLayerPerceptron)

```python
from tensorflow import keras
```

```python
# Create API
dnn = tf.keras.models.Sequential()
# Adding first hidden Layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding second hidden Layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

# adding output
dnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Train the model
dnn.compile(optimizer='adam', loss = 'binary_crossentropy', metrics =['accuracy'])

# apply training dataset and fit the model
dnn.fit(x_train, y_train, batch_size=32, epochs=100)
```

```python
dnn.fit(x_train, y_train, batch_size=32, epochs=100)
```

```python
dnn.summary()
```

### First Layer (`dense`)
- **Output shape**: `(None, 6)` — This means the layer outputs 6 neurons, with `None` representing the batch size (flexible input).
- **Parameters**: 72 — This suggests an input size of 11 (since \( 11 \times 6 + 6 = 66 + 6 = 72 \), accounting for weights and biases).

### Second Layer (`dense_1`)
- **Output shape**: `(None, 6)` — Another layer with 6 neurons.
- **Parameters**: 42 — This is calculated as \( 6 \times 6 + 6 = 36 + 6 = 42 \), indicating the input from the previous layer (6 neurons) plus biases.

### Third Layer (`dense_2`)
- **Output shape**: `(None, 1)` — A single output neuron, likely for a binary classification or regression task.
- **Parameters**: 7 — Calculated as \( 6 \times 1 + 1 = 6 + 1 = 7 \), based on the 6 inputs from the previous layer plus one bias.

### Total Parameters
- **121** (sum of all layer parameters: \( 72 + 42 + 7 \)).

```python
# Predict the test result
y_pred = dnn.predict(x_test)
```

```python
y_pred = (y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

**Class Imbalance:**

The dataset is heavily skewed toward class 0 (1585 vs. 415 samples). This imbalance likely causes the model to favor class 0, as seen in the high recall (0.97) and precision (0.88) for class 0, but poor recall (0.48) for class 1.

**Strengths:** 
The model is very good at identifying class 0 (high precision and recall), meaning it rarely misses class 0 instances and is confident when it predicts class 0.

**Weaknesses:**

The model struggles with class 1, with a low recall (0.48), meaning it misses more than half of the actual class 1 instances (false negatives). Precision for class 1 (0.79) is decent but not as strong as for class 0.

**Overall Performance:**

The accuracy (87%) looks good, but it’s inflated by the imbalance. The macro avg F1-score (0.76) gives a more balanced view, showing the model isn’t equally effective for both classes.

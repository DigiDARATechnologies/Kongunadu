```python
# '''scipy.spatial.distance- if a personâ€™s eyes are open or closed.

# imutils.face_utils - Helps in extracting eye, nose, and mouth landmarks from dlib's facial landmark detector.
# pygame- playing audio files.
# imutils - A Python library that provides convenient functions for image processing and working with OpenCV.
# dlib - pre-trained models for face and eye detection.
# cv2 - Captures video from a webcam.Detects faces and eyes in real-time.'''


from scipy.spatial import distance
from imutils import face_utils
from pygame import mixer
import imutils
import dlib
import cv2

# ''' Without calling mixer.init(), you cannot use the other audio functions. '''

mixer.init()
mixer.music.load("music.wav")

# '''This function calculates the Eye Aspect Ratio (EAR), which is a mathematical measure used to determine whether a person's eyes are open or closed.
# eye is an array containing six landmark points around the eye, detected using a facial landmark detector (dlib).
# A: The first vertical distance between landmark points 1 and 5.

# B: The second vertical distance between landmark points 2 and 4.

# C: The horizontal distance between landmark points 0 and 3.'''

def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear

# '''the code is setting up threshold values and initializing face detection and landmark prediction models using dlib.'''
# '''Variable Purpose
# thresh = 0.25    Sets the eye closure threshold for detecting drowsiness
# frame_check = 20   Ensures the eyes remain closed for 20 frames before triggering an alert
# detect = dlib.get_frontal_face_detector()  Detects faces in the video feed
# predict = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")  Detects 68 facial landmarks (including eyes, nose, and mouth)
#  The shape_predictor_68_face_landmarks.dat file is a pre-trained model used by Dlib to detect 68 facial landmarks (eyes, nose, mouth, jawline, etc.).'''
thresh = 0.25
frame_check = 20
detect = dlib.get_frontal_face_detector()
predict = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# '''The 68 facial landmarks model used in dlib assigns specific index numbers to key facial points. 
# Code    Purpose
# (lStart, lEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["left_eye"]    Gets the landmark indices for the left eye (36-41).
# (rStart, rEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["right_eye"]   Gets the landmark indices for the right eye (42-47).
# cap = cv2.VideoCapture(0)   Starts video capture from the webcam.
# flag = 0    Keeps track of consecutive frames with closed eyes.'''

(lStart, lEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["left_eye"]
(rStart, rEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["right_eye"]
cap=cv2.VideoCapture(0)
flag=0

# '''Code  Purpose
# cap.read()  Captures a video frame.
# imutils.resize()   Resizes the frame to 450 pixels width.
# cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)	Converts frame to grayscale for better processing.
# detect(gray, 0)  Detects faces in the frame.
# predict(gray, subject)  Finds 68 facial landmark points.
# shape[lStart:lEnd]   Extracts eye landmark points from the 68-point model.
# eye_aspect_ratio(eye)  Computes the Eye Aspect Ratio (EAR).
# cv2.convexHull(eye)  Creates a convex shape around the eye.
# cv2.drawContours()  Draws green eye contours on the frame.'''

while True:
    ret, frame=cap.read()
    frame = imutils.resize(frame, width=450)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    subjects = detect(gray, 0)
    for subject in subjects:
        shape = predict(gray, subject)
        shape = face_utils.shape_to_np(shape)
        leftEye = shape[lStart:lEnd]
        rightEye = shape[rStart:rEnd]
        leftEAR = eye_aspect_ratio(leftEye)
        rightEAR = eye_aspect_ratio(rightEye)
        ear = (leftEAR + rightEAR) / 2.0
        leftEyeHull = cv2.convexHull(leftEye)
        rightEyeHull = cv2.convexHull(rightEye)
        cv2.drawContours(frame, [leftEyeHull], -1, (0, 255, 0), 1)
        cv2.drawContours(frame, [rightEyeHull], -1, (0, 255, 0), 1)
#         '''CodePurpose
#             if ear < thresh:   Checks if the eyes are closed.
#             flag += 1   Counts how long the eyes stay closed.
#             if flag >= frame_check:  Confirms drowsiness if the eyes stay closed for too long.
#             cv2.putText(frame, "ALERT!")   Displays a warning message.
#             mixer.music.play()  Plays an alarm sound.
#             flag = 0    Resets the counter if the user opens their eyes.'''
        
        if ear < thresh:
            flag += 1
            print (flag)
            if flag >= frame_check:
                cv2.putText(frame, "****************ALERT!****************", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                cv2.putText(frame, "****************ALERT!****************", (10,325),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                mixer.music.play()
        else:
            flag = 0
            
# Code   Purpose
# cv2.imshow("Frame", frame)   Displays the video feed.
# cv2.waitKey(1) & 0xFF   Waits for a key press (without lagging).
# if key == ord("q"):   Exits the program when "q" is pressed.
# cv2.destroyAllWindows()   Closes all OpenCV windows.
# cap.release()   Releases the webcam.          
            
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
cv2.destroyAllWindows()
cap.release() 
```
